```python
import pandas as pd
import pyproj
import folium
```

```python
path = r"C:\Users\kiemd\Desktop\new map\Tê Giác Trắng\TGT.csv"
df = pd.read_csv(path)
df
```

```python
df['latitude'] = 1.1111
df['longitude'] = 1.1111
# Define the EPSG codes for the source and target CRS
src_crs = pyproj.CRS("epsg:32648")
ll_crs = pyproj.CRS("EPSG:4326")
#Function of converting
def transform(src_crs, ll_crs, point):
    # Create a transformer object to perform the transformation
    transformer = pyproj.Transformer.from_crs(src_crs, ll_crs)
    # Use the transformer to perform the transformation
    lat, lon = transformer.transform(*point)
    # Print the new coordinates (lat, long in EPSG 4326)
    print(f"New coordinates: ({lat:.6f}, {lon:.6f})")
    return round(lat, 7),round(lon, 7)

#Add field of lat/long after the transformer
for i in range(len(df)):
    lat, lon = transform(src_crs, ll_crs, (df.X[i], df.Y[i]))
    df['longitude'][i]=lon
    df['latitude'][i]=lat

#Save file to folder
df.to_csv(r"C:\Users\kiemd\Desktop\new map\Tê Giác Trắng\TGT_a.csv")
df
```

```python
# Create a Folium map centered on the first point in the DataFrame
map_center = [df['longitude'].iloc[0], df['latitude'].iloc[0]]
my_map = folium.Map(location=map_center, zoom_start=12)

# Add each point in the DataFrame as a marker to the map
for index, row in df.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']], 
                  popup=row['WELL_ID']).add_to(my_map)
# Display the map
my_map
```
