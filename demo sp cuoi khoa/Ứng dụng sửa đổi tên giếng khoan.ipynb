```python
from tkinter import *
from tkinter import messagebox
import pandas as pd
import openpyxl
from tkinter import filedialog
from tkinter import ttk

# tạo cửa sổ ứng dụng
window = Tk()

# setup thông tin cho window
window.title("Replace well name")
window.geometry("350x450")

# mô tả các function cho ứng dụng
def open_file_dialog():
    file_path = filedialog.askopenfilename()  # Mở hộp thoại chọn tệp
    entry_path.delete(0, END)  # Xóa nội dung hiện tại của Entry
    entry_path.insert(0, file_path)  # Đưa đường dẫn tệp vào Entry

def read_file():
    file_path = entry_path.get()  # Lấy đường dẫn tệp từ Entry
    if file_path:
        try:
            data = pd.read_excel(file_path)  # Đọc tệp Excel bằng Pandas
            # Sử dụng df cho mục đích xử lý dữ liệu
            print("Đọc tệp thành công.")
        except Exception as e:
            print("Lỗi khi đọc tệp:", str(e))
            messagebox.showinfo("Update Wellname", "Lỗi khi đọc tệp")
    else:
        print("Vui lòng nhập đường dẫn tệp trước khi đọc.")
        messagebox.showinfo("Update Wellname", "Vui lòng nhập đường dẫn tệp trước khi đọc.")
        
    # xử lý tên cũ trong file excel 
    # lấy cột tên giếng trong dataframe (từ excel)
    well_data = data.iloc[0: , 0]
    # Chuyển về dạng list 
    well_data_list = well_data.values.tolist()

    # lấy tên mới trong từ điển 
    # Lấy dữ liệu tên giếng từ file csv
    ten_gieng_da_sap_xep = open("tên giếng đã sắp xếp.csv", mode="r", encoding="utf-8-sig")
    # đưa các tên giếng đã sắp xếp vào một list chứa các list con
    dictionary = []
    while 1:
        row = ten_gieng_da_sap_xep.readline()
        if row is None or row == '': 
            break
        list_dict = row.split(",")
        new_name = list_dict[:2]
        dictionary.append(new_name)
    print(dictionary)
    
    # Loại bỏ ký tự "\n" từ các phần tử trong list con
    for sublist in dictionary:
        for i in range(len(sublist)):
            sublist[i] = sublist[i].rstrip("\n")
    
    # Tìm tên mới dựa vào tên cũ trong dictionary 
    result = []
    miss = []
    for item1 in well_data_list:
        found = 0
        for item2 in dictionary:
            if item1 == item2[1]:
                print(f"Chuỗi '{item1}' khớp với cặp giá trị {item2}")
                found = 1
                break
        
        # viết kết quả
        if found == 1:
            result.append(item2[0])
        else:
            print(f"ko tìm thấy cặp giá trị khớp với Chuỗi '{item1}'")
            miss.append(item1)
            result.append(item1)
    
    # Thêm dữ liệu vào TreeView từ một vòng lặp
    for i in miss:  # Hiển thị các giá trị từ 1 đến 10
        tree.insert("", "end", values=(i,))
    # Hiển thị TreeView
    tree.pack(padx=10, pady=10)
            
    # ghi kết quả ra một file excel mới 
    # Xóa cột đầu tiên 
    data = data.drop('UniqueId', axis=1)

    # Thêm cột mới vào lấy dữ liệu từ result
    data.insert(0, 'UniqueId', result)
    
    # In kết quả ra một tệp Excel
    data.to_excel('C:/Users/kiemd/Desktop/hocpython/hocpython/demo sp cuoi khoa/new_data1.xlsx', index=False) 
    # hiện thông báo 
    messagebox.showinfo("Update Wellname", "Đã sửa đổi tên giếng")

# setup các widget
# tạo entry
entry_path = Entry(window, width=50)
entry_path.pack(padx=10, pady=10)

# tạo button
button_open = Button(window, text="Chọn Tệp", command=open_file_dialog)
button_open.pack(pady=5)
button_read = Button(window, text="Đọc Tệp", command=read_file)
button_read.pack(pady=5)

# Tạo TreeView widget
tree = ttk.Treeview(window, columns=("Values",), show="headings")
tree.heading("Values", text="Các giếng không tìm thấy tên mới")

# chạy ứng dụng
window.mainloop()
```
