```python
def ve_nha(chieu_dai_nha,chieu_rong_nha):
    import turtle
    
    p = turtle.Turtle()
    p.pensize (6)
    p.speed(10)
    
    turtle.bgcolor("pink")
    p.pencolor("blue")
    p.fillcolor ("blue")
    
    p.begin_fill()
    
    # ve nha chinh
    quay_goc_vuong = 90
    p.forward(chieu_rong_nha)
    p.right(quay_goc_vuong)
    p.forward(chieu_dai_nha)
    p.right(quay_goc_vuong)
    p.forward(chieu_rong_nha)
    p.right(quay_goc_vuong)
    p.forward(chieu_dai_nha)
    
    p.end_fill()
    
    # nhac but len ve doi tuong moi
    p.penup()
    p.backward(120)
    p.pendown()
    
    p.pencolor("orange")
    p.fillcolor ("orange")
    
    p.begin_fill()
    # ve nha de xe
    p.left(90)
    p.forward(100)
    p.left(90)
    p.forward(80)
    p.left(90)
    p.forward(100)
    p.end_fill()
    
    p.penup()
    p.backward(15)
    p.pendown()
    
    p.pencolor("grey")
    p.fillcolor ("grey")
    
    p.begin_fill()
    # ve cua nha de xe
    p.left(90)
    p.forward(40)
    p.left(90)
    p.forward(70)
    p.left(90)
    p.forward(40)
    p.left(90)
    p.forward(70)
    p.end_fill()
    
    p.penup()
    p.forward(55)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("white")
    p.fillcolor ("brown")
    
    p.begin_fill()
    # ve cua chinh
    p.left(90)
    p.forward(50)
    p.right(90)
    p.forward(30)
    p.right(90)
    p.forward(50)
    p.right(90)
    p.forward(30)
    p.end_fill()
    
    p.penup()
    p.backward(10)
    p.right(90)
    p.forward(140)
    p.right(90)
    p.forward(5)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("white")
    p.fillcolor ("brown")
    
    p.begin_fill()
    # ve cua so
    p.circle(20)
    p.end_fill()
    
    p.penup()
    p.backward(250)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("black")
    p.fillcolor ("brown")
    
    p.begin_fill()
    # ve cay
    p.forward(10)
    p.right(90)
    p.forward(150)
    p.right(90)
    p.forward(10)
    p.right(90)
    p.forward(150)
    p.end_fill()
    
    p.penup()
    p.backward(30)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("green")
    p.fillcolor ("green")
    
                 
    p.begin_fill()
    # ve tan cay
    p.circle(30)
    p.end_fill()
    
    p.penup()
    p.right(90)
    p.forward(10)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("green")
    p.fillcolor ("green")
    
                 
    p.begin_fill()
    # ve tan cay
    p.right(90)
    p.circle(30)
    p.end_fill()
    
    p.penup()
    p.right(90)
    p.forward(5)
    p.right(90)
    p.forward(10)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("green")
    p.fillcolor ("green")
    
                 
    p.begin_fill()
    # ve tan cay
    p.right(90)
    p.circle(30)
    p.end_fill()
    
    p.penup()
    p.left(90)
    p.forward(200)
    p.right(90)
    p.forward(10)
    p.pendown()
    
    p.pensize (2)
    p.pencolor("yellow")
    p.fillcolor ("yellow")
    
    p.begin_fill()
    # ve mat troi
    ban_kinh_mat_troi = 40 
    p.right(90)
    p.circle(ban_kinh_mat_troi)
    p.end_fill()
    
    
    # ve tia nang
    p.right(90)
    p.penup()
    p.forward(20)
    p.pendown()
    
    for i in range(12):
     p.forward(20)
     p.backward(20)
     p.left(90)
     p.penup()
     p.circle(ban_kinh_mat_troi + 20,30)
     p.pendown()
     p.right(90)
    
    turtle.done()


ve_nha(chieu_dai_nha = 200,chieu_rong_nha = 100)
```
